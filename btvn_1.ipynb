```python
#23.6
import osmnx as ox

khu_vuc = "Thủ Đức, Thành Phố Hồ Chí Minh"
graph = ox.graph_from_place(khu_vuc, network_type='drive')

ox.plot_graph(graph)
thong_ke = ox.basic_stats(graph)

print(f" Số lượng nút giao: {thong_ke['n']}")
print(f" Số đoạn đường hiện có: {thong_ke['m']}")
print(f" Chiều dài trung bình của 1 con đường: {thong_ke['street_length_avg']:.1f} m")
```

```python
#23.11
import folium
from geopy.geocoders import Nominatim

may = Nominatim(user_agent="geoapi")

goc = may.geocode("Đại học Kinh tế Thành phố Hồ Chí Minh cơ sở B")
toa_do_goc = (goc.latitude, goc.longitude)

diem1 = may.geocode("Bệnh viện Chợ Rẫy")
diem2 = may.geocode("Hùng Vương Plaza")
diem3 = may.geocode("Sân vận động Thống Nhất")

du_lieu_nhu_cau = [
    (diem1.latitude, diem1.longitude, "Bệnh viện Chợ Rẫy", 85, "Đỏ"),
    (diem2.latitude, diem2.longitude, "Hùng Vương Plaza", 70, "Cam"),
    (diem3.latitude, diem3.longitude, "Sân vận động Thống Nhất", 30, "Xanh lá") ]

m = folium.Map(location=toa_do_goc, zoom_start=15)

for lat, lon, ten, chi_so, mau in du_lieu_nhu_cau:
    folium.Circle(
        location=[lat, lon],
        radius=chi_so * 3,
        color=mau,
        fill=True,
        fill_opacity=0.5,
        popup=f"{ten}: Dự báo {chi_so} lượt gọi xe/giờ"
 ).add_to(m)
m
```

```python
#23.12
import folium
from geopy.geocoders import Nominatim

may = Nominatim(user_agent="geoapi")
kho_hang = may.geocode("Đại học Kinh tế Thành phố Hồ Chí Minh cơ sở B")
toa_do_kho = (kho_hang.latitude, kho_hang.longitude)

diem_a = may.geocode("Bệnh viện Chợ Rẫy")
diem_b = may.geocode("Hùng Vương Plaza")
diem_c = may.geocode("Sân vận động Thống Nhất")

lo_trinh_toi_uu = [
    toa_do_kho,
    (diem_a.latitude, diem_a.longitude),
    (diem_b.latitude, diem_b.longitude),
    (diem_c.latitude, diem_c.longitude),
    toa_do_kho ]

m = folium.Map(location=toa_do_kho, zoom_start=15)
folium.Marker(toa_do_kho, popup="KHO HÀNG TRUNG TÂM").add_to(m)

folium.PolyLine(lo_trinh_toi_uu, color="green", weight=5, opacity=0.8, tooltip="Lộ trình tối ưu").add_to(m)

lo_trinh_xau = [toa_do_kho, (diem_c.latitude, diem_c.longitude), (diem_a.latitude, diem_a.longitude), toa_do_kho]
folium.PolyLine(lo_trinh_xau, color="gray", weight=2, opacity=0.5, dash_array='5, 5', tooltip="Lộ trình chưa tối ưu").add_to(m)
m
```
